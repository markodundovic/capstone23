# This next code reflext all the people worked on creating move, not only actors.


```python
df = pd.read_csv('MyFinalDataFilms1.csv')
```

```python
df.head()
```

```python
# Split the 'actors' column on the '-' delimiter, and convert to a list
df['credits'] = df['credits'].str.split('-')
```

```python
all_actors = []

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Get the list of actors for the current movie
    actors = row['credits']
    
    # Extend the list of all actors with the actors for the current movie
    all_actors.extend(actors)

# Convert the list of all actors to a set to remove duplicates
unique_actors = set(all_actors)

# Remove single-letter names from the set of unique actors
unique_actors = {actor for actor in unique_actors if len(actor) > 1}


# Print the total number of actors and the number of unique actors
print(f"Number of unique actors: {len(unique_actors)}")
```

```python
unique_actors
# Convert the set into a list
my_list = list(unique_actors)

# Create a dictionary with a single column
data = {'actors': my_list}

# Create a Pandas DataFrame from the dictionary
actors = pd.DataFrame(data)
actors.head()
```

```python
movie_df = df
actors_df= actors
```

```python
# Create an empty dictionary to store the total profits for each actor
actor_profits = {}
actor_counts = {}

# Loop over each row in the movie dataset
for index, row in movie_df.iterrows():
    # Get the profit and list of actors for the current movie
    profit = row["profit"]
    actors_list = row["credits"]
    # Loop over each actor in the list and add the profit to their total
    for actor in actors_list:
        if actor in actor_profits:
            actor_profits[actor] += profit
            actor_counts[actor] += 1
        else:
            actor_profits[actor] = profit
            actor_counts[actor] = 1

# Create a new list to store the total profits for each actor in the same order as the actors_df
total_profits = []
total_counts =[]
for actor in actors_df["actors"]:
    if actor in actor_profits:
        total_profits.append(actor_profits[actor])
        total_counts.append(actor_counts[actor])
    else:
        total_profits.append(0)
        total_counts.append(0)

# Add the new column of total profits to the actors dataset
actors_df["profit"] = total_profits
actors_df["count"] = total_counts

actors_df
```

```python
actors_df = actors_df.loc[actors_df['profit'] > 0]
```

```python
actors_df = actors_df.sort_values(by=['profit'], ascending=False)
```

```python
actors_df
```

```python
# Reset the index to default
actors_df = actors_df.reset_index(drop=True)

actors_df['profit_per_movie']=actors_df['profit']/actors_df['count']

# Format the profit values to be human-readable
actors_df['profit'] = actors_df['profit'].apply(lambda x: '${:,.2f}'.format(x))
```

# Genres top earnings


```python
df = pd.read_csv('MyFinalDataFilms1.csv')
```

```python
df.head()
```

```python
# Split the 'genres' column on the '-' delimiter, and convert to a list
df['genres'] = df['genres'].str.split('-')
```

```python
all_genres = []

for index, row in df.iterrows():

    genres = row['genres']
    
    all_genres.extend(genres)

unique_genres = set(all_genres)

unique_genres
```

```python
my_list = list(unique_genres)

data = {'genres': my_list}
genres = pd.DataFrame(data)
genres
```

```python
movie_df = df
genres_df= genres
```

```python
# Create an empty dictionary to store the total profits and count of movies for each genre
genre_profits = {}
genre_counts = {}

# Loop over each row in the movie dataset
for index, row in movie_df.iterrows():
    # Get the profit and list of genres for the current movie
    profit = row["profit"]
    genre_list = row["genres"]
    # Loop over each genre in the list and add the profit and count to their total
    for genre in genre_list:
        if genre in genre_profits:
            genre_profits[genre] += profit
            genre_counts[genre] += 1
        else:
            genre_profits[genre] = profit
            genre_counts[genre] = 1

# Create a new list to store the total profits and count of movies for each genre in the same order as the genres_df
total_profits = []
total_counts = []
for genre in genres_df["genres"]:
    if genre in genre_profits:
        total_profits.append(genre_profits[genre])
        total_counts.append(genre_counts[genre])
    else:
        total_profits.append(0)
        total_counts.append(0)

# Add the new columns of total profits and counts to the genres dataset
genres_df["profit"] = total_profits
genres_df["count"] = total_counts
```

```python
genres_df
```

```python
genres_df['profit_per_move']=genres_df['profit']/genres_df['count']
```

```python
genres_df = genres_df.sort_values(by=['profit_per_move'], ascending=False)
```

```python
# Reset the index to default
genres_df = genres_df.reset_index(drop=True)

# Format the profit values to be human-readable
genres_df['profit_per_move'] = genres_df['profit_per_move'].apply(lambda x: '${:,.2f}'.format(x))
```
