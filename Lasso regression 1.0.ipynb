```python
df = pd.read_csv('MyDataFilmsScaled.csv')
df.head()
```

```python
X = df[['budget', 'vote_average', 'popularity']]
y = df['revenue']
```

```python
model = Lasso(alpha = 20)
model
```

```python
model.fit(X,y)
```

```python
model.intercept_
model.coef_
```

```python
alphas = np.linspace(0.01, 500,1000)
alphas
```

```python
# Lasso regression coefficients
coefs = []
for alpha in alphas:
    #model fitting
    model = Lasso(alpha = alpha)
    model.fit(X, y)
    coefs.append(model.coef_)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([0, 50])
ax.plot(alphas, coefs)
ax.set_xlabel('alpha')
ax.set_ylabel('weights')
ax.set_title('Lasso coefficients as a function of the regularization');
```
