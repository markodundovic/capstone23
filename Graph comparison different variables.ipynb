```python
df = pd.read_csv('MyDataFilmsScaled.csv')
df.head()
```

```python
x2=np.log(df['budget'])
y2=np.log(df['revenue'])
df1 = pd.DataFrame({'budget_log': x2, 'revenue_log': y2})
```

```python
sns.regplot(x = 'budget_log', y = 'revenue_log', data = df1, scatter_kws={'alpha':0.5})
plt.title('Budget vs Revenue')          # Set title.
plt.xlabel("Budget ")       # Set x-axis label.
plt.ylabel('Revenue')    # Set y-axis label.
```

```python
sns.jointplot(x = 'budget_log', y = 'revenue_log', data = df1,alpha = 0.5)
```

```python
# Extract the data for the x and y axes
x_data = np.log(df[['budget', 'vote_average', 'popularity']])
y_data = np.log(df['revenue'])
```

```python

# Create a scatter plot
plt.scatter(x_data['budget'], y_data, label='budget')
plt.scatter(x_data['vote_average'], y_data, label='vote_average')
plt.scatter(x_data['popularity'], y_data, label='popularity')

# Add axis labels and a title
plt.xlabel('Budget, Vote Average, Popularity')
plt.ylabel('Revenue')
plt.title('Revenue vs. Budget, Vote Average, and Popularity')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
